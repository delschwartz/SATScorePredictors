# EDA on PSAT10 vs SAT demographic predictors of SAT scores

# Import packages & data

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
from scipy.stats import norm
```

```python
df = pd.read_csv('../data/SAT_PSAT_multicategories_v2.csv')
```

```python
df.info()
```

```python
# Fix dtypes after import

df = df.astype({'District Number': 'str',
                 'District Name': 'str',
                 'School Number': 'str',
                 'School Name': 'str',
                 'Median Family Income': 'float64',
                 'Population': 'float64'})
```

```python
df.info()
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True, errors='ignore')
```

```python
df.head()
```

```python
# Rename columns for convenience

df.columns = ['District Number', 'District Name', 'School Number', 'School Name',
       'Gend', 'Ethn', 'FRL', 'ELL', 'IEP', 'SAT Math',
       'SAT EBRW', 'Zip Code', 'Median Family Income', 'Population',
       'PSAT Math', 'PSAT EBRW']
```

# Continuous variable data exploration

```python
# Scores only with median family income

df_score_income = df[['SAT Math', 'SAT EBRW', 'PSAT Math', 'PSAT EBRW', 'Median Family Income']]
df_score_income.describe()
```

## All score distributions

```python
sns.set(font_scale=1)
plt.figure(figsize=(40,20))
sns.pairplot(df_score_income)
plt.show()
```

```python
sns.histplot(data=df, x='SAT Math', kde=True)
```

```python
sns.scatterplot(data=df, y='SAT Math', x='PSAT Math')
```

## PSAT vs SAT score distribution

```python
df1 = df.loc[:,['SAT Math', 'PSAT Math']]

axs= sns.jointplot('PSAT Math', 'SAT Math', data=df1)
axs.ax_joint.scatter('PSAT Math', 'SAT Math', data=df1)

axs.ax_marg_x.cla()
axs.ax_marg_y.cla()
sns.distplot(df1['PSAT Math'], ax=axs.ax_marg_x, fit=norm)
sns.distplot(df1['SAT Math'], ax=axs.ax_marg_y, vertical=True, fit=norm)
```

**Note:** 
* Score histograms for each test indicate possible normal distribution.
* Correlation implied between test scores & median family income

## Score correlation coefficients

```python
sns.set(font_scale=1.2)
plt.figure(figsize=(8,4))
plt.title('Heatmap of correlation coefficients')
sns.heatmap(df_score_income.corr(), annot=True)
plt.show()
```

### Observations

* High correlation between PSAT and SAT scores.  Implies that PSAT scores are good predictor of SAT scores.
* High correlation between different subject scores for each  year.  High math score $\Longleftrightarrow $ high EBRW score.
* Significant correlation between median family income & test scores.

# Categorical variables exploration

```python
df_scores = df[['SAT Math', 'SAT EBRW', 'PSAT Math', 'PSAT EBRW', 'Gend', 'Ethn', 'FRL', 'ELL', 'IEP', 'Median Family Income']]
df_scores.head()
```

```python
# One-hot encode categorical variables
col_to_onehotenc = ['Gend', 'Ethn', 'FRL', 'ELL', 'IEP']

encoder = OneHotEncoder(handle_unknown='ignore')
df_onehot = df_scores
for col in col_to_onehotenc:
    encoder_df = pd.DataFrame(encoder.fit_transform(df[[col]]).toarray())
    encoder_df.columns=encoder.get_feature_names_out()
    df_onehot = df_onehot.join(encoder_df)
    df_onehot = df_onehot.drop(columns=col)
```

```python
df_onehot.head()
```

```python
df_onehot.describe()
```

## Categorical variable correlation coefficients

```python
corr_coefs = df_onehot.corr()

sns.set(font_scale=2.5)
plt.figure(figsize=(40,20))
plt.title('Heatmap of correlation coefficients')

sns.heatmap(corr_coefs, annot=True)
plt.show()
```

```python
cc_small = corr_coefs.iloc[0:5,4:-1]
```

### Simplified heatmap

```python
sns.set(font_scale=2.5)
plt.figure(figsize=(40,10))
plt.title('Heatmap of correlation coefficients', fontsize='large')
sns.heatmap(cc_small, annot=True)
plt.show()
```

### Observations

Only first four columns (or first four rows) involve the target features.

**Noteworthy characteristics:**

* EBRW & Math mean scores (student scores for those subjects) are highly correlated.  High EBRW score $\Longleftrightarrow  $ high math score.

* Most important features for math score:
    * Free or reduced lunch (FRL)
    * Median family income (MFE)
    * Ethnicity (Hispanic)
    * Engligh language learner (ELL) status
    * Ethnicity (white)
    
* Significant segregation between white & hispanic students may be implied by -0.91 correlation between those two groups.
* Expected inverse correlation between FRL and MFE is observed.  Higher median income $\Longleftrightarrow  $ fewer students qualifying for FRL.

```python
# Get total for each category.
df_onehot.drop(['SAT Math', 'SAT EBRW', 'PSAT Math', 'PSAT EBRW', 'Median Family Income'], axis=1).sum()
```

**Noteworthy characteristics:**

* Few groups of students with IEPs, may not be enough for statistical significance.  However these are groups of students, not individuals.  Scores were only reported if there were at least 17 students in a subgroup.  This may not matter since IEP status had a very low correlation coefficient.

```python
df_onehot.to_csv('../data/scores_onehot.csv')
```

# Major takeaways from initial EDA

* PSAT10 scores in 2017 are highly correlated with SAT scores in 2018.  Correlation coefficient 0.97 for both EBRW and Math scores.
* There are 5 demographic variables which each have a magnitude greater than 0.5 with SAT scores in 2018.

**The big question** is whether the most significant demographic variables collectively outweigh the correlation between PSAT and SAT scores.

It is important to note that the same demographic variables impact PSAT and SAT scores in the same way.  The question here involves a veil of ignorance.  If we *only* know the PSAT scores will that be a better predictor than if we *only* know the demographic variables?
